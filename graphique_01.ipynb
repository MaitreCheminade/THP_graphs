# Premier essai graphique 
Réaliser un graphique avec deux axes. L'axe des abscisses va de 0 à 50 et s'intitule "Axe des x". L'axe des ordonnées s'intitule "Axe des y".
Tracer la droite d'équation : x = 3y. Avec x allant de 0 à 50.
Le graphe possède le titre suivant "Tracer une droite" et la droite est de couleur rouge.

 ***

```python
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
# x est espacé de 0 à 50, en 100 points 
x = np.linspace(0, 50, 100)
# valeur de y ou équation de la courbe 
y = x / 3 
# on indique qu'on veut un graphique et les données qu'on y met en X et en Y, et la couleur rouge 
plt.plot(x, y, color = "r")
# les labels 
plt.xlabel("Axe des x")
plt.ylabel("Axe des y")
plt.title("Tracer une droite")
plt.show()
```

```python
# avec seaborn
import seaborn as sb 
import matplotlib 
# Créer un graphique avec Seaborn
sb.lineplot(x=x, y=y, color = "r")

# Personnaliser les étiquettes des axes
import matplotlib.pyplot as plt
plt.xlabel("Axe des x")
plt.ylabel("Axe des y")

# Afficher le graphique
plt.show()
```
